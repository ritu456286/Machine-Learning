```python
# Import necessary libraries 
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA 
from sklearn.datasets import load_iris 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
# Load the Iris dataset 
iris = load_iris() 
X = iris.data # Feature data (4 features) 
y = iris.target # Class labels (3 classes) 
```

```python
X.shape
```

```python
# Initialize LDA and reduce to 2 components 
lda = LDA(n_components=2) 
 
lda
```

```python
# Fit and transform the data 
X_lda = lda.fit_transform(X, y)
```

```python
X_lda.shape
```

```python
# Create a DataFrame for better visualization 
lda_df = pd.DataFrame(data=X_lda, columns=['Component 1', 'Component 2']) 
lda_df['Target'] = y 
```

```python
lda_df
```

```python
# Visualize the LDA results 
plt.figure(figsize=(8, 6)) 
colors = ['r', 'g', 'b'] 
labels = ['Setosa', 'Versicolour', 'Virginica'] 
 
for target, color, label in zip([0, 1, 2], colors, labels): 
    plt.scatter(lda_df[lda_df['Target'] == target]['Component 1'], 
    lda_df[lda_df['Target'] == target]['Component 2'], color=color, label=label) 
 
plt.title('LDA on Iris Dataset') 
plt.xlabel('Component 1') 
plt.ylabel('Component 2') 
plt.legend() 
plt.grid() 
plt.show()
```
