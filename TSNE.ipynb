```python
# Import necessary libraries 
from sklearn.manifold import TSNE 
from sklearn.datasets import load_iris 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
# Load the Iris dataset 
iris = load_iris() 
X = iris.data # The feature data (4 features) 
y = iris.target # The class labels (3 classes) 
```

```python
X.shape
```

```python
# Initialize t-SNE with 2 components 
tsne = TSNE(n_components=2, random_state=42)
tsne
```

```python
# Fit and transform the data 
X_tsne = tsne.fit_transform(X) 
```

```python
X_tsne.shape
```

```python
# Create a DataFrame for better visualization 
tsne_df = pd.DataFrame(data=X_tsne, columns=['Dimension 1', 'Dimension 2']) 
tsne_df['Target'] = y 
```

```python
# Visualize the t-SNE result 
plt.figure(figsize=(8, 6)) 
colors = ['r', 'g', 'b'] 
labels = ['Setosa', 'Versicolour', 'Virginica'] 
 
for target, color, label in zip([0, 1, 2], colors, labels): 
    plt.scatter(tsne_df[tsne_df['Target'] == target]['Dimension 1'], 
    tsne_df[tsne_df['Target'] == target]['Dimension 2'], color=color, label=label) 
 
plt.title('t-SNE on Iris Dataset') 
plt.xlabel('Dimension 1') 
plt.ylabel('Dimension 2') 
plt.legend() 
plt.grid() 
plt.show()
```
