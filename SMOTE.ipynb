```python
# Import necessary libraries 
from sklearn.datasets import make_classification 
from imblearn.over_sampling import SMOTE 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# Create an imbalanced dataset 
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, 
                           n_redundant=0, n_clusters_per_class=1, 
                           weights=[0.9, 0.1], flip_y=0, random_state=42)
```

```python
X.shape, y.shape
```

```python
X
```

```python
y
```

```python
# Visualize the original class distribution 
print("Original class distribution:", np.bincount(y))
```

```python
# Plot the imbalanced dataset before applying SMOTE 
plt.figure(figsize=(10, 5)) 
 
plt.subplot(1, 2, 1) 
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolor='k') 
plt.title("Before SMOTE (Imbalanced)") 
plt.xlabel("Feature 1") 
plt.ylabel("Feature 2")
```

```python
# Initialize SMOTE 
smote = SMOTE(random_state=42)
```

```python
smote
```

```python
# Apply SMOTE to balance the dataset 
X_resampled, y_resampled = smote.fit_resample(X, y) 
```

```python
X_resampled.shape
```

```python
y_resampled.shape
```

```python
# Visualize the new class distribution after SMOTE 
print("New class distribution after SMOTE:", np.bincount(y_resampled))
```

```python
# Plot the balanced dataset after SMOTE 
plt.subplot(1, 2, 2) 
plt.scatter(X_resampled[:, 0], X_resampled[:, 1], c=y_resampled, cmap='coolwarm', 
edgecolor='k') 
plt.title("After SMOTE (Balanced)") 

plt.xlabel("Feature 1") 
plt.ylabel("Feature 2") 
 
plt.tight_layout() 
plt.show() 
```
