```python
# Import necessary libraries 
from sklearn.decomposition import PCA 
from sklearn.datasets import load_iris 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
# Load the Iris dataset 
iris = load_iris() 
X = iris.data # The feature data (4 features) 
y = iris.target # The class labels (3 classes) 
```

```python
X.shape
```

```python
# Initialize PCA with 2 components 
pca = PCA(n_components=2) 
```

```python
pca
```

```python
# Fit and transform the data 
X_pca = pca.fit_transform(X) 
```

```python
X_pca.shape
```

```python
# Create a DataFrame for better visualization 
pca_df = pd.DataFrame(data=X_pca, columns=['Principal Component 1', 'Principal Component 2']) 
pca_df['Target'] = y
```

```python
# Visualize the 2D PCA results 
plt.figure(figsize=(8, 6)) 
colors = ['r', 'g', 'b'] 
labels = ['Setosa', 'Versicolour', 'Virginica']
for target, color, label in zip([0, 1, 2], colors, labels): 
    plt.scatter(pca_df[pca_df['Target'] == target]['Principal Component 1'], 
    pca_df[pca_df['Target'] == target]['Principal Component 2'], 
    color=color, label=label) 
 
plt.title('PCA on Iris Dataset') 
plt.xlabel('Principal Component 1') 
plt.ylabel('Principal Component 2') 
plt.legend() 
plt.grid() 
plt.show()
```

```python
#Explained variance ratio 
print("Explained variance ratio of the 2 principal components:", 
pca.explained_variance_ratio_) 
```
