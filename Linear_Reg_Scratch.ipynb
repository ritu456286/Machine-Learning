```python
import numpy as np

# Generate some sample data for illustration
np.random.seed(42)
X = 2 * np.random.rand(100, 1)  # Feature matrix (100 samples, 1 feature)
y = 4 + 3 * X + np.random.randn(100, 1)  # Target variable with some noise

# Add a column of ones to X for the bias (intercept) term
X_b = np.c_[np.ones((100, 1)), X]  # Add bias term to X

# Initialize parameters (weights)
theta = np.random.randn(2, 1)  # Two parameters: one for bias, one for the feature

# Hyperparameters
learning_rate = 0.1
n_iterations = 1000
m = len(X_b)  # Number of samples

# Gradient Descent
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)  # Compute gradients
    theta -= learning_rate * gradients  # Update parameters

# Display final parameters
print(f"Intercept: {theta[0][0]}, Coefficient: {theta[1][0]}")

# Predict for new data
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]  # Add bias term
y_pred = X_new_b.dot(theta)

print("Predictions for new data:", y_pred)

# Optional: Plot the results
import matplotlib.pyplot as plt

plt.scatter(X, y, color='blue', label='Data')
plt.plot(X_new, y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
